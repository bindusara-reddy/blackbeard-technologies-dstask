```python
import os
import numpy as np
import mne
from matplotlib import pyplot as plt
from scipy.fft import fft,fftfreq
import pandas as pd
import glob
```

```python
raw_data = mne.io.read_raw_edf('edf data/ST7011J0-PSG.edf')
annot_data = mne.read_annotations('edf data/ST7011JP-Hypnogram.edf')
mapping = {'EOG horizontal': 'eog', 'EMG submental': 'emg','Marker': 'misc'}
raw_data.set_annotations(annot_data)
raw_data.set_channel_types(mapping)

print(raw_data.info)
```

```python
raw_data.plot(start=3000, duration=500,scalings=dict(eeg=2e-4, eog=4e-4, emg=2e-4,misc=1e-4))
```

```python
sampling_freq = raw_data.info['sfreq']
N=50 # no of seconds
start_stop_seconds = np.array([0, N])
start_sample, stop_sample = (start_stop_seconds * sampling_freq).astype(int)
print(start_sample,stop_sample)
channel_index = 0
raw_selection = raw_data[channel_index, start_sample:stop_sample]
print(raw_selection)
```

```python
yf= np.abs(fft(raw_selection[0][0]))
xf= fftfreq(int(N*sampling_freq),(1/sampling_freq))
plt.plot(xf,yf)
scalex=1
plt.xlim(min(xf)*scalex, max(xf)*scalex)
scaley=1.2
plt.ylim(min(yf)*scaley, max(yf)*scaley)
plt.show()
```

```python
signal_array=[]
hypnogram_array=[]
for name in glob.glob('edf data/*-PSG.edf'):
    signal_array.append(name)
for name in glob.glob('edf data/*-Hypnogram.edf'):
    hypnogram_array.append(name)
```

```python
for i in range(20):
    
    scale=1e-3
    
    temp_raw_data=mne.io.read_raw_edf(signal_array[i])
    temp_annot_data = mne.read_annotations(hypnogram_array[i])
    print(signal_array[i],hypnogram_array[i])
    mapping = {'EOG horizontal': 'eog', 'EMG submental': 'emg','Marker': 'misc'}
    temp_raw_data.set_annotations(temp_annot_data)
    temp_raw_data.set_channel_types(mapping)
    eeg_channels = mne.pick_types(temp_raw_data.info, eeg=True,eog=False,emg=False,misc=False)
    temp_raw_data.plot(duration=60, order = eeg_channels, proj=False,
                  n_channels=len(eeg_channels), remove_dc=False,scalings = scale)
    
    
    raw_bandpass = temp_raw_data.copy().load_data().filter(l_freq=0.25, h_freq=10)
    fig = raw_bandpass.plot(duration=60, order=eeg_channels, proj=False,
                            n_channels=len(eeg_channels), remove_dc=False,scalings = scale)
    fig.subplots_adjust(top=0.9)
    print('Band-pass filtered between 0.25 and 10 Hz')
```
